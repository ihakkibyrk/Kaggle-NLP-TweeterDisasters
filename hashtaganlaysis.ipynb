```python
import pandas as pd
import re
import nltk
```

```python
# Load the data
df =pd.read_pickle('clean_train.pkl')
```

```python
# Hashtag
def hashtag(text):
    hashtag_list = re.findall(r'#\w+', text)
    return hashtag_list
df['msg_hashtag'] = df['text'].apply(hashtag)
```

```python
tag = list(df.msg_hashtag[df['msg_hashtag'].apply(len)>0])
```

```python
#removing hashtags having numeric value & of lenght 1-2 letter
hashtags = []
for i in tag:
    for j in i:
        if bool(re.match(r'#\d+',j)) == False and len(j) > 3:
            hashtags.append(j)

# removo unnecessary hashtags
# hash_to_remove = ['#youtn','...']
# hashtags = [hashtag for hashtag in hashtags if hashtag not in hash_to_remove]
```

```python
hash_count = {}
for i in hashtags:
    hash_count[i] = hashtags.count(i)

# Display the df
hashtg = pd.DataFrame.from_dict(hash_count, orient='index')
hashtg.rename(columns={'index':'hashtags',0:'count'}, inplace=True)
hashtg.sort_values(by='count', ascending=False).head(30)
```
